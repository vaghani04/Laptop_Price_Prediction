```python
dict = {}
for label,content in df.items():
    if pd.api.types.is_string_dtype(content):
        dict[label] = df[label].unique()
dict
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ['Company',
'TypeName',
'ScreenResolution',
'Cpu',
'Memory',
'Gpu',
'OpSys']




one_hot = OneHotEncoder()
transformer = ColumnTransformer([('one_hot', one_hot, categorical_features)],
                                remainder='passthrough'
                               )
transformed_X_test = transformer.fit_transform(X)
transformed_X_test
```

```python
pd.DataFrame(transformed_X_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def predict_vs_actual_data(true_data, pred_data, dataframe, color="red"):
    sns.lmplot(x = true_data, y = pred_data, data = dataframe, fit_reg = False, height = 5, aspect = 2)
    d_line = np.arange(dataframe.min().min(), dataframe.max().max())
    # print(d_line, d_line.dtype)
    # temp = pd.DataFrame(d_line)
    # print(temp)
    plt.plot(d_line, d_line, color = color, linestyle = '--')
    plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

def evaluate_metrics(y_true, y_pred):
    r2 = r2_score(y_true = y_true, y_pred = y_pred)
    mae = mean_absolute_error(y_true = y_true, y_pred = y_pred)
    mse = mean_squared_error(y_true = y_true, y_pred = y_pred)
    rmse = np.sqrt(mse)
    return pd.DataFrame({"r2_score":[r2],
            "mean_absolute_error":[mae],
            "mean_squared_error":[mse],
            "root_mean_squard_error":[rmse]
            },)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(transformed_X_test, y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Model-1: Linear Regeression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
lr_accuracy_train = lr.score(X_train,y_train)
lr_accuracy_test = lr.score(X_test,y_test)
lr_accuracy_train, lr_accuracy_test
```

```python
lr_predict = lr.predict(X_test)
lr_predict[:10]
```

```python
lr_dataframe = pd.DataFrame({'actual_data': y_test, 'predicted_data': lr_predict})
lr_dataframe
```

```python
predict_vs_actual_data(true_data = 'actual_data', pred_data = 'predicted_data', dataframe = lr_dataframe)
```

```python
lr_metrics = evaluate_metrics(y_true = y_test, y_pred = lr_predict)
lr_metrics
```

## Model-2: Linear Regression (Ridge)

```python
from sklearn.linear_model import Ridge

l2 = Ridge()
l2.fit(X_train, y_train)
l2_accuracy_train = l2.score(X_train, y_train)
l2_accuracy_test = l2.score(X_test, y_test)
l2_accuracy_train, l2_accuracy_test
```

```python
l2_predict = l2.predict(X_test)
l2_dataframe = pd.DataFrame({'actual_data': y_test, 'predicted_data': l2_predict})
predict_vs_actual_data(true_data = 'actual_data', pred_data = 'predicted_data', dataframe = l2_dataframe)
```

```python
l2_metrics = evaluate_metrics(y_true = y_test, y_pred = l2_predict)
l2_metrics
```
